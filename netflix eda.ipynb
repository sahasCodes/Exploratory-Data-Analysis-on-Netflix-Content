Q2: Who are the top 10 actors on Netflix?


```python
#Fillinf NAN with values
df['cast']=df['cast'].fillna("No cast mention")

#Creating a Data frame to store the filtered cast(i.e Actors)

cast_df=pd.DataFrame()
cast_df=df["cast"].str.split(',',expand=True).stack()
cast_df=cast_df.to_frame()
cast_df.columns=["Actor"]
actors=cast_df.groupby(['Actor']).size().reset_index(name='Total Content')
actors=actors[actors.Actor != 'No cast mention']

#Sorting the top 10 

actors=actors.sort_values(by=['Total Content'],ascending=False)
top_actors=actors.head(10)
top_actors=top_actors.sort_values(by=['Total Content'])
x= top_actors['Actor']
y=top_actors['Total Content']

#Ploting the Bar graph
plt.style.use("fivethirtyeight")
plt.figure(figsize = (17,8))
plt.title('Top 10 Actors')
sns.barplot(x = x, y = y)
plt.show()
```

Q3: Does Netflix upload more Movies or TV shows



```python
plt.figure(figsize=(15,8))

plt.plot(df[df['type']=="TV Show"].groupby("release_year")['type'].count(),color="#E50914",linewidth=4)
plt.plot(df[df['type']=="Movie"].groupby("release_year")['type'].count(),color="black",linewidth=4)

plt.xlabel("Year")
plt.ylabel("Titles Added")
plt.title("Movies VS TV Shows released on Netflix per Year")
plt.legend(['TV Shows','Movies'])
plt.show()
```

 The chart shows the historical trends in the release of TV Shows and movies on Netflix.
Looking at the lines representing TV shows and movies, both show an upward trajectory, suggesting an overall increase in the quantity of content being added to Netflix's library.

Specifically, the line representing TV shows red appears to have a steeper incline, indicating a relatively higher rate of growth compared to movies,represented by the black line. This suggests that Netflix has been focusing on expanding its TV show offerings more rapidly than its movie catalog.

Overall, the trend suggests that Netflix has been consistently adding more TV shows and movies to its platform over time, reflecting the company's efforts to diversify and expand its content library to cater to a wide range of viewerÂ preferences.
